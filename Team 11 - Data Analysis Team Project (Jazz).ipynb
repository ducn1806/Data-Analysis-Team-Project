# *Introduction - Southern and Eastern Africa*

The cultures within Southern and Eastern Africa are multi-layered. Both regions are rich and diverse in culture, ethnicity, language, cuisine, geography, social and religious practices, and more. Social hierarchies are very significant in these regions; elders and people in positions of authority are well-respected. Thousands of languages spoken in Eastern/Southern Africa such as Swahili, Zulu, Shona, and Xhosa.
 
Geographically, South Africa is situated in the middle of the Indian and South Atlantic oceans. East Africa is off the coast of the Gulf of Aden and the Arabian sea. The Great Escarpment, the Kalahari desert and the Zambezi river are three of the most prominent  geographic features in Southern Africa. Eastern Africa is home to Mount Kilimanjaro, Mount Kenya as well as famous national parks and game reserves like the Serengeti and Masai Mara .
 
After the end of apartheid, which is a system of  institutionalised racial segregation that existed in South Africa from 1948 until the early 1990s, employment increased. However, due to restrictions and discrimination against indigenous Africans, more than half of people who wanted to work moved into self-employment, also categorized as informal employment. Indeed, recently revised time series of Statistics South Africa suggest that between 1994 and 2000, informal labour has spread widely.  
 
Like all regions of the world, Eastern and Southern Africa have their fair share of socio-economic issues. Since 1994, life expectancy has fallen by almost 20 years â€“ mainly because of the rise in HIV-related mortality. In 2006, life expectancy at birth in South Africa was 50.1 years, which is rather low, although still the highest in the sub-Saharan region. The expectancy was 51.0 years for females and 49.1 percent for male. The prevalence of these issues in both regions guided our research and approach to this analysis.


# *Analysis*

We noticed that although Togo is listed in our region, it is actually a West African country, so we decided to remove it as an **obscure** value before moving forward. 


We selected the five features that best describe our region through a muli-step process. First, we classified the columns in our region into five categories of economy, technology, health, environment, and politics. Second, we filtered out the columns with more than one missing value, because too many missing values would impact the integrity of our analysis, in accordance with the rare event rule.

```python
### importing packages
import pandas            as pd
import matplotlib.pyplot as plt 
import seaborn           as sns
import numpy             as np

# the path to the dataset
file = "Final_Project_Dataset.xlsx"

# reading the file into Python through pandas
full = pd.read_excel(file)

# filtering for Jazz region
jazz      = full.loc[ : , 'Cool Name' ] == 'Jazz'
# filtering for the rest of the world
rest      = full.loc[ : , 'Cool Name' ] != 'Jazz'
# filtering for Sideswipe region
sideswipe = full.loc[ : , 'Cool Name' ] == 'Sideswipe'
# filtering for Bumblebee region
bumblebee = full.loc[ : , 'Cool Name' ] == 'Bumblebee'
# removing Togo from filter
rem_togo  = full.loc[ : , 'Country Name' ] != 'Togo'

# filtering the region into 5 categories
# filter for the category economy
economy     = full[jazz][rem_togo].iloc[ : , [10, 11, 12, 13, 14, 15, 18, 19,\
                                              20, 25, 42]]

# filter for the category health
health      = full[jazz][rem_togo].iloc[ : , [4, 6, 7, 8, 17, 21, 24, 27, 29,\
                                              31, 35, 36, 37, 40, 44]]

# filter for the category environment
environment = full[jazz][rem_togo].iloc[ : , [9, 16, 22, 23]]

# filter for the category technology
technology  = full[jazz][rem_togo].iloc[ : , [26, 30]]

# filter for the category politics
politics    = full[jazz][rem_togo].iloc[ : , [39]]


# makes the new DataFrame independent for economy
economy_filter = pd.DataFrame.copy(economy)
    
# for loops to find columns with only a maximum of 1 missing value
for col in economy_filter:

    # if statement to look for columns that has more than 1 missing value
    if economy_filter[col].isnull().sum(axis = 0) > 1:
        # deletes the column if there is more than 1 missing value
        del economy_filter[col]
        
# makes the new DataFrame independent for health      
health_filter = pd.DataFrame.copy(health)
    
# for loops to find columns with only a maximum of 1 missing value
for col in health_filter:

    # if statement to look for columns that has more than 1 missing value
    if health_filter[col].isnull().sum(axis = 0) > 1:
        # deletes the column if there is more than 1 missing value
        del health_filter[col]

# makes the new DataFrame independent for environment
environment_filter = pd.DataFrame.copy(environment)
    
# for loops to find columns with only a maximum of 1 missing value
for col in environment_filter:

    # if statement to look for columns that has more than 1 missing value
    if environment_filter[col].isnull().sum(axis = 0) > 1:
        # deletes the column if there is more than 1 missing value
        del environment_filter[col]

# makes the new DataFrame independent for technology
technology_filter = pd.DataFrame.copy(technology)
    
# for loops to find columns with only a maximum of 1 missing value
for col in technology_filter:

    # if statement to look for columns that has more than 1 missing value
    if technology_filter[col].isnull().sum(axis = 0) > 1:
        # deletes the column if there is more than 1 missing value
        del technology_filter[col]

# makes the new DataFrame independent for politics        
politics_filter = pd.DataFrame.copy(politics)

# for loops to find columns without missing values
for col in politics_filter:

    # if statement to look for columns that has more than 1 missing value
    if politics_filter[col].isnull().sum(axis = 0) > 1:
        # deletes the column if there is more than 1 missing value
        del politics_filter[col]


# filter for top 5 features for Jazz w/o Togo
top5 = full[jazz][rem_togo].iloc[ : , [1, 23, 26, 36, 39, 42]]
```

# *Correlation Between Top 5 Features*

We checked for the correlation between the five selected variables.  The strongest correlations were between HIV prevalence and water source to the total number of self-employed (-0.62 and -0.65 respectively). However, these factors are not particularly related and we know that correlation does not equal causality. The next highest correlation is between internet users and the number of self-employed people. The majority of self-employed people in the region are engaged in informal, and often unskilled labour, which does not require internet access (van Klaveren, et al. 2020). This could explain why there is a relatively high correlation of -0.56 between both factors.

```python
#######################################
###     Heatmap for Top 5 Columns   ###
#######################################

# converting correlation matrix into a DataFrame
top5_corr = top5.corr(method = 'pearson').round(decimals = 2)

# instantiating a heatmap
fig, ax = plt.subplots(figsize=(10,10))


# developing a spicy heatmap
sns.heatmap(data       = top5_corr,     # the correlation matrix
            cmap       = 'inferno',     # changing to SPICY colors
            square     = True,          # tightening the layout
            annot      = True,          # should there be numbers in the heatmap
            linecolor  = 'black',       # lines between boxes
            linewidths = 0.5)           # how thick should the lines be?


# title and displaying the plot
plt.title("""
Linear Correlation Heatmap for Top 5 Features
""")

# displaying the plot
plt.tight_layout()
plt.show()

############################################
###     Scatter plot for Top 5 Columns   ###
############################################

# instantiating a scatter plot for internet users and self-eployment
sns.lmplot(x          = 'Internet users (per 100 people)',  # x-axis feature
           y          = 'Self-employed, total (% of total employment)',  # y-axis feature
           hue        = None,     # categorical data for subsets
           legend_out = False,    # formats legend if hue != None
           scatter    = True,     # renders a scatter plot
           fit_reg    = True,     # renders a regression line
           aspect     = 2,        # aspect ratio for plot
           data       = top5)     # DataFrame where features exist


# formatting and displaying the plot
plt.xlabel(xlabel  = 'Internet users (per 100 people)')
plt.ylabel(ylabel  = 'Self-employed, total (% of total employment)')
plt.xlim(0, 10)  # changing the x-range for better viewability
plt.ylim(0, 100) # changing the y-range for better viewability

# displaying the plot
plt.tight_layout()
plt.show()
```

3. **HIV**: When comparing the mean from East and Southern Africa with the rest of the world, we can see results that are consistent with our research. 4. **Seats**: On average, East/Southern Africa is doing better than the rest of the world in terms of number of seats held by women in parliament.
